### C)

### D)

```python
import torch
```

```python
grouped_by_userId = merged_df.groupby(["userId"]).groups
```

### E)

```python
from sklearn.preprocessing import normalize
```

```python
n_users = ratings_df.userId.max().item()
n_movies = movies_df.shape[0]
train_X = torch.zeros((n_users, n_movies))
train_X.shape
```

```python
for i in range(n_users):
    indices = grouped_by_userId[i+1]
    movie_indices = merged_df.iloc[indices]["List Index"].to_numpy() - 1
    rates = torch.tensor(merged_df.iloc[indices]["rating"].to_list())
    train_X[i, movie_indices] = rates
```

```python
train_X = torch.tensor(normalize(train_X, axis=1), dtype=torch.float32)
```

### F)

first define the network

```python
import torch.utils.data
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable
from torchvision import datasets, transforms
from torchvision.utils import make_grid , save_image
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
class RBM(nn.Module):
   def __init__(self, n_vis=784, n_hin=500):
        super(RBM, self).__init__()
        self.W = nn.Parameter(torch.randn(n_hin,n_vis)*1e-2)
        self.v_bias = nn.Parameter(torch.zeros(n_vis))
        self.h_bias = nn.Parameter(torch.zeros(n_hin))
        self.sigmoid = nn.Sigmoid()
        self.relu = nn.ReLU()
        
    
   def sample_from_p(self,p):
       return self.relu(torch.sign(p - torch.rand(p.size())))
    
   def v_to_h(self,v):
        p_h = self.sigmoid(torch.mm(v, self.W.t())+self.h_bias)
        return p_h
    
   def h_to_v(self,h):
        p_v = self.sigmoid(torch.mm(h,self.W)+self.v_bias)
        return p_v
        
   def forward(self,v):
        h1 = self.v_to_h(v)
        v1 = self.h_to_v(h1)
        
        return v,v1
    
   def free_energy(self,v):
        vbias_term = v.mv(self.v_bias)
        wx_b = F.linear(v,self.W,self.h_bias)
        hidden_term = wx_b.exp().add(1).log().sum(1)
        return (-hidden_term - vbias_term).mean()

```

### G)

```python
rbm = RBM(n_vis=n_movies, n_hin=20)
train_op = optim.SGD(rbm.parameters(),0.1)
for epoch in range(20):
    loss_ = []
    sample_data = train_X.bernoulli()
    
    v,v1 = rbm(sample_data)
    loss = (rbm.free_energy(v) - rbm.free_energy(v1))
    loss_.append(loss.data)
    train_op.zero_grad()
    loss.backward()
    train_op.step()

    print(F"Training loss for {epoch+1} epoch: { np.mean(loss_)}")
```

### H)

```python
user75_15_top_movies_indices = rbm.forward(train_X)[1][75].sort().indices[0:15]
```
